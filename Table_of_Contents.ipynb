These set of notebooks query the Cloud Pak for Data Cluster. The first notebook demonstrates various SQL statements running against virtualized data sources. The second notebook will populate the Dashboard server with summary information on a number of stocks. The final notebook will run through each stock in the database and populate the Dashboard continuously.
